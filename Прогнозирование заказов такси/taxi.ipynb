#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

### Наблюдается тренд роста заказов за период наблюдения.  Наблюдается сезонность по дням.  Спад заказов ночью, рост заказов с утра и днем, пик заказов такси к вечеру.


## Обучение

#### Создадим фунуцию для  создания признаков.

```python
def make_features(data, max_lag, rolling_mean_size):
    data['year'] = data.index.year
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    data.dropna(inplace=True)
```

#### Создадим признаки.

```python
make_features(df, 6, 3)
df.sample(1)
```

#### Разделим данные на тренировочную и тестовую выборки.

```python
features = df.drop(['num_orders'], axis=1) # выделение признаков
target = df['num_orders'] # целевая переменная
```

```python
features_train, features_test, target_train, target_test = train_test_split(
    features, target, shuffle=False, test_size=0.1, random_state=12345)
print(features_train.shape)
print(features_test.shape)
```

#### Создадим и обучим различные модели. Рассчитаем метрику RMSE на обучающей выборке. 

```python
def RMSE(target,predictions):
    MSE = mean_squared_error(target, predictions)
    return MSE ** 0.5
RMSE_score = make_scorer(RMSE, greater_is_better=False)
```

```python
tscv = TimeSeriesSplit(n_splits=5)
```

#### Линейная регрессия

```python
%%time
model = LinearRegression()
scores = cross_val_score(model, features_train, target_train, scoring = RMSE_score, cv = tscv)
print(scores)
```

#### DecisionTreeRegressor

```python
%%time
model_DT = DecisionTreeRegressor(random_state=12345)
param_DT = {'max_depth':range (1,10)}
grid_DT = GridSearchCV(model_DT, param_DT, scoring = RMSE_score, cv = tscv)
grid_DT.fit(features_train,target_train)

print(grid_DT.best_params_)
print(grid_DT.best_score_)
```

#### RandomForestRegressor

```python
%%time
model_RF = RandomForestRegressor(random_state=12345)
param_RF = {'n_estimators':[100, 150],
             'max_depth':range (3, 20, 3)}
grid_RF = GridSearchCV(model_RF, param_RF, scoring = RMSE_score, cv = tscv)
grid_RF.fit(features_train,target_train)

print(grid_RF.best_params_)
print(grid_RF.best_score_)
```

#### LightGBM

```python
%%time
model_lgbm = LGBMRegressor(random_state=12345)
param_lgbm = {
    'n_estimators': [100, 200, 500],
    'max_depth': [15,30],
    'num_leaves': [20, 50, 100]
}
grid_lgbm = GridSearchCV(model_lgbm, param_lgbm, scoring = RMSE_score, cv = tscv)
grid_lgbm.fit(features_train,target_train)

print(grid_lgbm .best_params_)
print(grid_lgbm .best_score_)
```

#### Результаты 

```python
table = pd.DataFrame([(-1)*scores.mean(), (-1)*grid_DT.best_score_, (-1)*grid_RF.best_score_, (-1)* grid_lgbm .best_score_], 
             columns=['RMSE'], 
             index=['LR','DecisionTrees', 'RandomForest','LGBM' ])
display(table)
```

### Из итоговой таблицы можно сделать вывод, что  лучшая модель - это LGBM. 

## Тестирование

#### Проведем тестирование и посчитаем метрику RMSE для нашей лучшей модели на  тестовой выборке.

```python
predict = grid_lgbm.predict(features_test)
print('RMSE модели LGBM на  тестовой выборке', RMSE(target_test, predict))
```

```python
def graph(target, pred):
    sns.set(rc={'figure.figsize':(15,4)})
    result = target.to_frame().reset_index()
    result['prediction'] = pd.Series(pred)
    result.set_index('datetime', inplace=True)
    result.plot()
    plt.show()
```

```python
graph(target_test, predict)
```
